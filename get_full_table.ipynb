```python
import pandas as pd
df = pd.read_csv('constitutions_noindex_noeconding.csv')
df
```

```python
df['country'] = df['link']
df['country'] = df['country'].map(lambda x: x.lstrip('https://www.constituteproject.org/constitution/').rstrip('?lang=en'))
df['year'] = df.country.str.extract('(\d+)')
df['country'] = df.country.replace('\d+', '', regex=True)
df['country'] = df['country'].map(lambda x: x.rstrip('_'))
df = df[['link', 'country', 'year', 'name', 'text']]
df = df.sort_values(by=['country'])
df = df.reset_index(drop=True)
df
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
df.to_csv('full_table.csv', index = False)
```
